### Heatmap:
    Uzun vadeli zaman serisine dayalı veriler olduğunda bu görselleştirme yapılabilir.
    Bu dönemlere karşı gözlemlenebilecek sayısal değerler olduğunda.
    Ya da daha büyük olcekli belirli periyotlarla tekrar eden olaylar olduğunda bu görselleştirme kullanılabilir.
    Çok sınıflı veriler varsa onlarında görselleştirmede işimize yaramaktadır.

```python
# Kütüpanenin import edilmesi.
import seaborn as sns

# Kütüpane içinde olan hazır veri setinin kullanılması.
flights = sns.load_dataset("flights")


# Veri setinde yapısal bir değişiklik olmadan veri setinin bir kopyasının alınması.
df = flights.copy()

# Veri setinin ilk beş verisinin yazdırlıması.
df.head()
```

```python
# Veri setinin son beş satırına bakıyorum.
df.tail()
```

```python
# Verisetinde ilgili sürtünlarda boş veri seti varmı ona bakıyoruz.
df.isnull().sum()
```

```python
# Veri setinin veri yapılarına bakıyoruz.
df.dtypes
```

```python
# Veri setindeki verilerin istaiksel betimlerine bakıyoruz.
df.describe().T
```

```python
# Veri setinin şekline bakıyorum.
df.shape
```

```python
# Burda "passengers" adlı sütündaki verilerin ile ilgili istaiksel değerlere bakıyorum.
df["passengers"].describe()
```

Veri setini pivot bir veri türünde olacak şekilde işleme tabi tutmazsak bu veri heatmap gibi işlemlere tabi tutamayız.

```python
df = df.pivot(index="month", columns="year", values="passengers") # Anahtar argümanları vermeden bu kod çalışmaz ve hata alırsınız.
```

```python
df.head()
```

```python
sns.heatmap(df);
```

```python
Isı haritaları elimizde yapı saptamak gibi problemler olduğunda oldukça işimize yararlar.
```

```python
# Farklı görselleştirme biçimleri.

# Hücreleri kendi değerleri ile doldurma ve görselleştirme.
sns.heatmap(df, annot = True, fmt = "d");
```

```python
# Hücre aralarına çizgi çekelim.

sns.heatmap(df, annot = True, fmt = "d", linewidths = .5);
```

```python
# Sol tarftaki barı kaldırmak istediğimizde is şu kodu yazmalıyız.

sns.heatmap(df, annot = True, fmt = "d", linewidths = .5, cbar = False);
```

```python
import matplotlib.pyplot as plt

# Isı Haritası
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap="YlGnBu", linewidths=0.5, fmt = "d")
plt.title("Passenger Counts Heatmap")
plt.show()
```
