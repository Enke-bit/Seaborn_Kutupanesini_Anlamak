```python
# Veri setinin import edilmesi.

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Kütüpanedeki hazır olan veri setini kullanma.

tips = sns.load_dataset("tips")
```

```python
# Veri setin ilk beş verisini görelim.

tips.head()
```

```python
# Veri setinde yapısal bir değişiklik yapmadan veri setinin bir copyasını alma.

df = tips.copy()
```

```python
# Kopsayı alınan veri setinin ilk veri satırını yazdırma.

df.head()
```

Violin grafiği nasıldır biraz baxplot grafiğine benxer ama yoğunlukta içerir bu iki grafiğin kesişimi gibidir.

```python
# Çıkaçak olan uyarıları kapatalım.

import warnings

# Uyarıları devre dışı bırak
warnings.filterwarnings('ignore', category=UserWarning, module='seaborn.axisgrid')
```

```python
sns.catplot(y = "total_bill", kind = "violin", data = df);
```

Violinde Çaprazlamalar:

```python
sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df);
```

```python
# Bir kırılım ekleyelim. (kırılım = boyut anlamında kullanılıyor.)

sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df, hue = "sex");
```

```python
# İki violin grafiğini tek bir violin grafiğinde kullanmak isterseniz şu kullanımı yapabilirsiniz.

sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df, split = True);
```

```python
# Violin grafiğine farlı iç grafikler çizebilir.

sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df, inner = "quartile");
```

```python
# Grafiklerin prizlülüğünü ayarlayabilirsiniz.

sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df, bw = 0.2);
```

```python
# Grafiğin genişliğini yoğunluğuna göre ölçütleyebilirsiniz.

sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df, scale = "count");
```

```python
# Grafikte yoğunluk hesaplama noktalarını yazdırablirsiniz.

sns.catplot(x ="day", y = "total_bill", kind = "violin", data = df, gridsize = 40);
```

```python
# farklı renkte renk paleti kullanablirsiniz.

sns.catplot(x = "day", y = "total_bill", kind = "violin", data = df, hue = "sex", palette = "pastel");
```
