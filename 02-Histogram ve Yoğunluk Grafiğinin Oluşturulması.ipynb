```python
# Gerekli kütüpanelerin import edilmesi.

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Gerekli olan veri setinin yüklenmesi.

diamonds = sns.load_dataset('diamonds')
```

```python
# Veri setinde yapısal değişiklik olmayan halinin bir kopyasını alıyorum.

df = diamonds.copy()
```

```python
df.head()
```

```python
# Dağılım görselleştirme için kullanılacak olan grafik ayarlanmasını yapıyorum.

sns.displot(df.price, kde = False);          # kde yoğunluk grafiğinin fonksyona koyulup koyulmaması ile ilgili

# Bu garfik sürekli değişken ile ilgili bize gerekli olan bilgiyi vermektedir.
```

```python
# Renk atmak istersem.

sns.displot(df.price, kde = False, color = "purple");
```

```python
#Orentısal histogram kullanmak istersem.

sns.displot(df.price, kde=False, stat='density');
```

```python
# Yoğunluk eğrisi eklemek istedim.

sns.displot(df.price, kde=True);
```

```python
sns.displot(df.price, kde = True, color = 'blue');
```

```python
# Grafiğe başlık eklemek.
sns.displot(df.price, kde = False);
plt.title('Fiyat dağılımı')
plt.show()
```

```python
# X ve Y eksenlerine isim verilebilir.

sns.displot(df.price, kde = False)
plt.xlabel('Fiyat dağılımı')
plt.ylabel('Fiyat aralığı dağılımı')
plt.show()
```

```python
# Stil kullanımı

sns.displot(df.price, kde = True, linestyle = '--');
```

```python
# Şefaflık ayarlaması.

sns.displot(df.price, kde = True, alpha = 0.3)
```

```python
# Birden fazla değşşekn için dağılım yapmak istersem.
# Hue ile gruplama yapmak zorındayım.

sns.displot(df, x = 'price', hue = 'cut', kind = "hist", alpha = 0.25);
```

```python
# Çoklu grafik tanımlaması.

sns.displot(df, x = "price", col = "cut");

# Bu içerdeki farklı özellikleri farklı grafikler ile göstermeyi sağlamaktadır.
```

```python
# Çoklu grafiklerin düzenlenmesi.

sns.displot(df, x = "price", col = "cut", col_wrap=3);
```

```python
# Sadece yoğunluk grafiğini görmek istersek.

sns.displot(df.price, kind="kde");
```

```python
# Yoğunluğun içini dolu olarak bakamazmıyız.

sns.kdeplot(df.price, shade = True);
```

- Bu grafiklerin her ikiside sayısal grafiklerin dağılması için kullanılacağı için her ikiside yada sadece biri kuyllanılabilir.
- Sürekli değişkenlerin dağılımlarını görsellştirmede önemli rol oynamaktadır.

## Çapraz gösterim:

```python
# Şimid süreklü veri gösterimi ile grafik oluştur dediğimizde seaborn bize bu şekilde bir gösterim sağlamaktadır.

sns.kdeplot(df.price, shade = True);
plt.title("Fiyatların 'price' sürekli gösterimi ile gösteren tablo");
plt.ylabel("Frekansların standartlaştırılmış hali");
plt.xlabel("Fiyat gösterimi");
```

Biz çaprazlama ile bu grafik nasıl olmuş onun  hakkında bilgi sahibi olmaktayız.

Şimid eğer derinlik verilmemiş olan grafikte iki tepeli bir grafik varsa allta aslında birdan fazla türde veri olduğu anlamına gelmektedir.

```python
# Çıkan uyarılar fazla olduğu için ekledim.
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

(sns
 .FacetGrid(df, hue = "cut", height=5, xlim = (0, 10000))
 .map(sns.kdeplot, "price", shade = True)
 .add_legend()
);
```

```python
# Daha farklı bir gösterimle ele alacak olursak.

sns.catplot(x = "cut", y = "price", hue = "color", kind = "point", data = df);
```
