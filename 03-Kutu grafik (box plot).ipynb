## Boxplot
Veri Seti Hikayesi
- total bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)
- tip: bahşiş
- sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)
- smoker: grupta sigara içen var mı? (0=No, 1=Yes)
- day: gün (3-Thur, 4-Fri, 5=Sat, 6=Sun)
- time: ne zaman? (0-Day, 1=Night)
- size: grupta kaç kişi var?

Sayısal bir değişkenin dağılımı incelenmek isteniyorsa histogram grafikle birlikte kutu grafiğininde incelenmeasi lazım.

```python
import seaborn as sns             # Seaborn kütüpanesinin import etmeye yönelik kodumu yazıyorum.
tips = sns.load_dataset("tips")   # Seaborn kütüpanesindeki hazır data setlerden biri olan bir hazır veri setini çağırıyorum.
df = tips.copy()                  # Veri setinin yapısal bir değişklik olmadan önce bir kopyasını alıyorum.
df.head()                         # Veri setinin ilk beş satırını sütünları ile birlikte yazdırıyorum.
import matplotlib.pyplot as plt 
```

### Boxplot görselleştirme işlemine başlama.

```python
# Total yemeğe gelen insanların ödediği hesap.

sns.boxplot(x = df["total_bill"]);
```

```python
# Yemeğe gelen insanların bıraktığı tip.

sns.boxplot(x = df["tip"]);
```

### Grafik çaprazlama işlemi.

1. Hangi günler daha fazla kazaniyoruz.

```python
sns.boxplot(x = "day", y = "total_bill", data = df);
print(df["day"].value_counts())

# Yukardaki soruya burdan bazı cevaplar bulabiliriz.
```

```python
# Kişi sayısını bi boyut olarak ekleyelim.

sns.boxplot(x = "day", y = "total_bill", data = df, palette="viridis", hue = "size");
```

2. Sabahmı akşam mı daha çok kazanıyoru?

```python
sns.boxplot(x = "time", y = "total_bill", data = df);
```

```python
# Kategorilere göre renkleri ayarlama

sns.boxplot(x="time", y="total_bill", data=df, palette="Set2")
```

```python
# Özel istatiksel verileri göstercek şekilde görselleştirme.

sns.boxplot(x = "time", y = "total_bill", data = df, showfliers = False);
```

```python
# Daha kapsayıcı şekilde görselleri alma.

sns.boxplot(x = "time", y = "total_bill", data = df, notch = True);
```

```python
# Çizgi şeklinde görselleştirmesini yapılması.

sns.boxplot(x="time", y="total_bill", data=df, linewidth=2.5)
```

```python
# Dikey şeklinde görselleştirmesi yapılması.

sns.boxplot(x = "total_bill", y = "time", data = df, orient="h")
```

```python
# Grafiklerin üzerine sayısal değerlerini girmne.

sns.boxplot(x = "time", y = "total_bill", data = df);

sns.scatterplot(x = "time", y = "total_bill", data = df, color = "red", marker = "o", s = 50, label = "Ortalama");
plt.legend()
plt.show()
```

```python
# Kişi sayısı ile kazanç oranı gösterimi.

sns.boxplot(x = "size", y = "total_bill", data = df);
```

```python
# gün ie total hesap ve üçüncü boyut ile cinsiyeti ele alalım.

sns.boxplot(x = "day", y = "total_bill", hue = "sex", data = df);
```

```python
# Kişi sayısını bi boyut olarak ekleyelim.

sns.boxplot(x = "time", y = "total_bill", data = df, hue = "size");
```

```python
# Bu sefer kişilerin gelidiği günlerden bir grafiğe çaprazlama verdim.

sns.boxplot(x = "time", y = "total_bill", data = df, hue = "day");
```
