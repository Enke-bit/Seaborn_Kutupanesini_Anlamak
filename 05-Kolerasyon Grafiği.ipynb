#### Koleraksyon ne demek:
    
    - Değişkenler arasındaki iliştiki ifade eder.
    - İstatiksel bir terimdir.
    - İki değişken arasındaki bağlantıyı ifade etmekte en fazla kullanılan grafik 'scatterplot' grafiğidir.
    
#### scatterplot:
    
    - Saçılım grafiği yaklaşımıdır.
    - Sayısal değişkenler arasındaki ilişkiyi göstermektedir.
    - Bu farklı değişkenler arasındaki ilişkilri inceleyerek bize veri seti hakkındaki yapı hakkında bilgi vermektedir.
   
   
#### Genel:
    
    - tek değişkenli veri görselleştirmeden iki değişkenli veri görselleştirmeye yönelik bir çalışma olduğunu söylemek istiyorum ve artık iki değişkein bir sayısal yapı üzerindeki etkisinin nasıl olduğunu anlamaya yönelik bir çalışma bu.
    - Burda yeni boyutlar ekleyerek veri arkassındaki bilgiye ulaşmaya yönelik çalışmalarda olucak.
    
    

#### Veri Seti Hikayesi

- total bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)
- tip: bahşiş
- sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)
- smoker: grupta sigara içen var mı? (0=No, 1=Yes)
- day: gün (3-Thur, 4-Fri, 5=Sat, 6=Sun)
- time: ne zaman? (0-Day, 1=Night)
- size: grupta kaç kişi var?

```python
# Veri setinin import edilmesi.
import seaborn as sns

#Kullanaçağımız hazır veri setinin tanımlanması.
tips = sns.load_dataset("tips")

# Veri setininde yapısal bir değişiklik yapmadan önce veri setinin bir kopyasını almak.
df = tips.copy()

# Veri setinin ilk beş satırını yazdırma.
df.head()
```

```python
# İki adet sürekli değişkenin dağılımına yönelik bir görselleştirme yapıyorum.

sns.scatterplot(x = "total_bill", y = "tip", data = df);
```

### Çaprazlamalar:

```python
# Veriye time ile boyut katalım.

sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data = df);
```

- Veri dayalı karar alma işlemini yapılmasında önemli rol oynar.
- Veri odağının istatiksel çalışmalar üzerinden çalışmaya dayalı işlem yapmasına dayanır.
- Kolerasyon ve ve çaprazlama ile bu dağılımların gözlemlenmesini sağlayabilir.

```python
# Şimid time katogorik değişeknin özeliği üzerinden işlem yapmaya dayalı işlem yapmaya devam ediyor.

sns.scatterplot(x = "total_bill", y = "tip", hue = "time", style = "time", data = df);
```

```python
# Farklı argümalar deniyorum.

sns.scatterplot(x = "total_bill", y = "tip", hue = "day", style = "time", data = df);
```

```python
# Peki bir sürekli değişkeni boyut olrak eklemek istersek nasıl işlem yapmamız gerekmektedir.

sns.scatterplot(x = "total_bill", y = "tip", size = "size", data = df);

# Sürekli ve sayısal olan bir değişkeni ben boyut özeliğini göz önünde bulundurarak görselleştirme yapıyorum.
```

```python
# Yukardaki kodda biraz daha okunaklığa yönelik bir katman katayım.

sns.scatterplot(x = "total_bill", y = "tip", hue = "size", size = "size", data = df);
```

```python
# Scatterplot oluşturma
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="size",
    size="sex",
    palette="viridis",
    sizes=(20, 200),
    alpha=0.6,
    edgecolor="w",
    legend="full",
    data=df  # Verinin doğru şekilde sağlandığından emin olun
);
```

### Scatterplot Matrisi

Bu ifade şu anlama gelmektedir bu saçılım grafiği üzerinden bu grafiğin oluşan matriksinin ele alınıp açıklanması olarak ele almak gerekmektedir.

```python
# veri seti yükleme

iris = sns.load_dataset("iris")

# Veri setinde yapısal değişiklik olmadan veri setinin kopyasını alma.

df = iris.copy()

# Veri setinin ilk beş verisini yazdıralım.

df.head()
```

```python
df.tail()
```

```python
# Veri seti veri yapılarına bakalım.

df.dtypes
```

```python
# Veri setinin şekline bakalım.

df.shape
```

```python
# Veri seti içindeki dört değişkenin birbiri araındaki ilişkisine bakalım.

sns.pairplot(df);

# Eğer bu görselleştirme sonuçunda oluşan grafikler toz bulutu ise bir anlam çıkmıyorsa o ii değişken arasında bir ilişki yoktur anlamına gelir.
# Eğer bu gafik veriler farklı köşelerde gruplanıyorsa burda bazı alt gruplar vardır ve onların varlığını bize ifade eder.
```

```python
# Şimdi bir boyut ekliyorum ve türleri ifade eden değişkeni kullanıyorum.

sns.pairplot(df, hue = "species");
```

```python
# Bu markerlama yani işaretleme türlerini değiştirebilirz.

sns.pairplot(df, hue = "species", markers = ["s","o","d"]);
```

```python
# Grafiğe doğru ekleme işlemi.

sns.pairplot(df, kind = "reg");
```

```python
# Bu eğimlere birde türleri ekleyelim.

sns.pairplot(df, kind = "reg", hue = "species");
```
