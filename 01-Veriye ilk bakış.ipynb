#### Veri Seti Hikayesi
- price: dolar cinsinde fiyat (326-18,823)
- carat: ağırlık (0.2-5.01)
- cut: kalite (Fair, Good, Very Good, Premium, Ideal)
- color: renk (from J (worst) to D (best))
- clarity: temizliği, berraklığı (11 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (BEST))
- x: length in mm (0-10.74)
- y: width in mm (0-58.9)
- z: depth in mm (0-31.8)
- depth: toplam derinlik yüzdesi = z / mean(x, y) = 2* z/ (x + y) (43-79)
- table: elmasin en geniş noktasına göre genişliği (43-95)

```python
# VERİ SETİ HAZIR BE İLGİLİ KÜTÜPANE İCİNDE BULLUNMAKTA BU YÜZDEN İMPORT İŞLEMLERİ SIRASINDA İNDİRİYORUM.

import seaborn as sns                     # İlgili kütüpanen import işlemi.
diamonds = sns.load_dataset('diamonds')   # hazır veri setinin kütüpaneden getirilmesi.
df = diamonds.copy()                      # Veri setinin bir kopyasının alınma işlemi.
df.head()                                 # Veri setinin ilk beç veri satırını yazdırma işlemi.
```

Veri setindeki bu katogorik değişkenler kaliteyi ifade ediyor.

Veri setim özelinde veriler en kötüden en iyiye doğru sıralnması söz konusu o yüzden bu işlemlerde ben bunu python programlama dilinde ifade etmem gerekecek.

```python
# ordinal tanımlama.

from pandas.api.types import CategoricalDtype
```

```python
df.cut.head()
```

```python
# Şimidi cut isimli değişekenin içindeki index yürlerini ordinal bir şekilde tanımlama işlemlerini gerçekleştirelim.

df.cut = df.cut.astype(CategoricalDtype(ordered=True))
```

```python
df.dtypes
```

```python
df.cut.head()
```

```python
cut_katogoriler = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
```

```python
# Şimid yukardaki doğru tanımlama işlemine göre tekrar işlem yaptırıyorum.

df.cut = df.cut.astype(CategoricalDtype(categories = cut_katogoriler, ordered=True))
```

```python
df.cut.head(1)
```

Sütün grafiği 'bar plot' grafiği oluşturma.

```python
df["cut"].value_counts().plot.barh().set_title("Cut değişkenin sınıf frekansları grafiği");
```

```python
# Kodların daha okunaklı olmasını sağlamak amaıyla yapılacak olan düzenleme.

(df["cut"]
 .value_counts()
 .plot.barh()
 .set_title("Cut değişkenin sınıf frekansları grafiği"));
```
